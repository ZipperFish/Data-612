Social media apps such as Facebook, Instagram, and Twitter use likes on a post to recommend content. TikTok relies heavily on watch time to understand whether a user paused a video, rewatched it, or spent only a few seconds watching it. Some users may watch a video without liking, sharing, or commenting, but still spend a minute or more engaging with the content. Short videos tend to receive more attention from the recommendation system, as the system learns from the amount of time users spend on TikTok. In addition to watch time, scroll speed indicates user interest, since fast scrolling suggests disinterest. The time of day a user engages with TikTok is also taken into account, as the feed adapts differently for content consumed at night versus in the morning.


TikTok's machine learning system is based on a **hybrid filtering approach**. With content-based filtering, the system analyzes video features such as captions, hashtags, audio, and visual elements. With collaborative filtering, it uses patterns from users with similar viewing behavior to recommend videos. After TikTok finishes filtering these videos, the system begins to rank them to keep the user engaged, focusing on watch time, completion rates, and shares. Each video is assigned a score based on engagement signals, and the highest-ranking videos are shown on the user's feed. The system continues adapting based on the user's interest profile and preferences.


### Attacks on Recommender System


Video games can suffer a significant impact from recommendation systems, as gamers come from many backgrounds and expect different stories, characters, and plots from a game. Recently, an upcoming video game in the *God of War* series, titled *God of War: Laufey*, has led some gamers to organize review bombing against the game due to a female protagonist being the main character. Unfortunately, video game companies struggle to control online discourse and how their games are received due to major disagreements about changes in their products.

To prevent such attacks, a video game platform such as Steam can use detection systems to identify when many new accounts suddenly start downvoting a video game or when there is repetitive activity that appears to be automated. Authentication systems can also help filter out fake accounts that influence results and limit how much a single user can impact the ranking of a game. Some platforms require users to purchase or download a video game before they can leave a rating. Platforms that use graph-based analysis can examine relationships between users and content to detect tightly connected groups that may be planning to manipulate ratings. However, I think one of the best ways to prevent such attacks is to apply trust-weighted modeling, where interactions from older and verified users have more influence on video game ratings.
